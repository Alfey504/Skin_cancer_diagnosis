```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from keras.utils.np_utils import to_categorical
```

```python
features = pd.read_pickle("loaded_data/features.pkl")
```

```python
target = pd.read_pickle("loaded_data/target.pkl")
```

```python
features.head()
```

```python
target.head()
```

```python
x_train_o, x_test_o, y_train_o, y_test_o = train_test_split(features, target, test_size=0.20,random_state=1234)
```

```python
x_train_o.head(15)
```

```python
x_test_o.head(15)
```

```python
y_train_o.head(15)
```

```python
y_test_o.head(15)
```

```python
x_train = np.asarray(x_train_o['image'].tolist())
x_test = np.asarray(x_test_o['image'].tolist())
```

```python
x_train_mean = np.mean(x_train)
x_train_std = np.std(x_train)

print(x_train_mean)
print(x_train_std)
```

```python
x_test_mean = np.mean(x_test)
x_test_std = np.std(x_test)
```

```python
x_train = (x_train - x_train_mean)/x_train_std
x_test = (x_test - x_test_mean)/x_test_std
```

```python
y_train = to_categorical(y_train_o, num_classes = 7)
y_test = to_categorical(y_test_o, num_classes = 7)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
x_train, x_validate, y_train, y_validate = train_test_split(x_train, y_train, test_size = 0.1, random_state = 2)
```

```python
x_train = x_train.reshape(x_train.shape[0], *(75, 100, 3))
x_test = x_test.reshape(x_test.shape[0], *(75, 100, 3))
x_validate = x_validate.reshape(x_validate.shape[0], *(75, 100, 3))
```

```python
np.save("loaded_data/x_train.npy", x_train,  allow_pickle=True)
np.save("loaded_data/y_train.npy", y_train,  allow_pickle=True)
```

```python
np.save("loaded_data/x_validate.npy", x_validate,  allow_pickle=True)
np.save("loaded_data/y_validate.npy", y_validate,  allow_pickle=True)
```

```python
np.save("loaded_data/x_test.npy", x_test,  allow_pickle=True)
np.save("loaded_data/y_test.npy", y_test,  allow_pickle=True)
```
